```python
# Подключение библиотек
import matplotlib.pyplot as plt  
import pandas as pd  
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder 
from datetime import datetime
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.pipeline import Pipeline, make_pipeline
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
data_train = pd.read_csv('./data/train.csv')
data_test = pd.read_csv('./data/test.csv')
data_sample_submission = pd.read_csv('./data/sample_submission.csv')
pd.set_option('display.max_columns', None)
```

```python
print('Выведем данные по обучающим данным')
display(data_train.head(10))
data_train.info()
```

```python
# удалим пустые значения которые нельзя заменить логически
data_train = data_train.dropna(subset=['transmission', 'make', 'model', 'color', 'condition', 'trim', 'body'])
data_train.info()
```

```python
display(data_train.loc[data_train['odometer'].isna()])
```

У машин 2014 года заменим NaN в столбце odometer на 0, так как они могут быть новые, потому что объявления выложены в 2015 году, остальные же удалим

```python
data_train.loc[(data_train['odometer'].isna()) & (data_train['year'] == 2014), 'odometer'] = 0
data_train = data_train.dropna(subset='odometer')
data_train.info()
```

```python
#Преобразование даты
data_train['saledate'] = pd.to_datetime(data_train['saledate'], utc=True)
data_train.info()
```

```python
data_train = data_train.loc[(data_train['color'] != '—') & (data_train['interior'] != '—')]
```

```python
# Избавимся от ненужных в исследовании столбцов
data_train = data_train.drop('vin', axis=1)
```

```python
#Добавим возраст машины при выставлении объявления и удалим ставщие ненужными столбцы с годом выпуска машиы и даты ваставления
# объявления
data_train['car_age'] = data_train['saledate'].dt.year - data_train['year']
data_train = data_train.drop(['year','saledate'], axis=1)
```

```python
#Поссмотрим корреляцию Пирсона и двустороннее значение p в обучающих данных
print(data_train.corr())  
    
# plotting the correlation heatmap  
data_train_plot = sns.heatmap(data_train.corr(), cmap="YlGnBu", annot=True)  
    
# displaying the heatmap  
plt.show()
```

```python
cols = ['make', 'model', 'trim']
data_train['full_auto_name'] = data_train[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
data_train = data_train.drop(cols, axis=1)
display(data_train.head(10))
```

```python
encoder = OrdinalEncoder()
columns_need = ["body","transmission", "state", "color", "interior", "seller", "full_auto_name"]
encoder.fit(data_train[columns_need])
data_train[columns_need] = encoder.transform(data_train[columns_need])
display(data_train)
```

```python
features_train, target_train = data_train.drop(columns = 'sellingprice', axis=1), data_train['sellingprice']
```

```python
display(features_train, target_train)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=12345)
params = {
    'max_depth': [8, 10, 12, 14],
}

imba_pipeline = make_pipeline(RandomOverSampler(random_state=12345), 
                              DecisionTreeClassifier(random_state=12345))

new_params = {'decisiontreeclassifier__' + key: params[key] for key in params}
best_model_DTC = GridSearchCV(imba_pipeline, param_grid=new_params, cv=kf, scoring='recall',
                        return_train_score=True)

best_model_DTC.fit(features_train, target_train)

print(best_model_DTC.best_params_)
print(best_model_DTC.best_score_)
```
