```python
# Подключение библиотек
import matplotlib.pyplot as plt  
import pandas as pd 
from datetime import datetime
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from catboost import CatBoostRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.model_selection import GridSearchCV
```

```python
data_train = pd.read_csv('./data/train.csv')
data_test = pd.read_csv('./data/test.csv')
data_sample_submission = pd.read_csv('./data/sample_submission.csv')
pd.set_option('display.max_columns', None)
```

```python
print('Выведем данные по обучающим данным')
display(data_train.head(10))
data_train.info()
```

```python
print('Выведем данные по тестовым данным')
display(data_test.head(10))
data_test.info()
```

```python
#Превратим столбец saledate в дату вместо строки
def time_for_age(data):
    data['saledate'] = pd.to_datetime(data['saledate'], utc=True)
    data['car_age'] = data['saledate'].dt.year - data['year']
    data = data.drop(['year','saledate'], axis=1)
    return data

data_train = time_for_age(data_train)
data_test = time_for_age(data_test)
```

```python
#В тренировочных данных удалим значения NaN, так как заменить их логически мы не сможем
data_train = data_train.dropna(subset=['transmission', 'make', 'model', 'color', 'condition', 'trim', 'body'])
# В значениях с пробегом у машин с пробегом NaN и 0 возрастом можем поставить пробег 0 так как возможно пробег не указан,
# потому что машины новые
data_train.loc[(data_train['odometer'].isna()) & (data_train['car_age'] == 0), 'odometer'] = 0
data_test.loc[(data_test['odometer'].isna()) & (data_test['car_age'] == 0), 'odometer'] = 0
data_train = data_train.dropna(subset='odometer')
data_train.info()
#для тестовых данных все NaN заменим на -
for i in data_test.loc[data_test['odometer'].isna(), 'car_age'].unique():
    data_test.loc[(data_test['odometer'].isna()) & (data_test['car_age'] == i), 'odometer'] = data_test.loc[data_test['car_age'] == i, 'odometer'].median()
data_test.loc[data_test['condition'].isna()] = data_test['condition'].median()
data_test = data_test.fillna(value='—')
```

```python
print('Проверка дубликатов:', data_train['vin'].str.lower().duplicated().sum())
```

```python
#Проверим аномальные значения и не явные дубликаты
print(sorted(data_train['make'].unique()))
```

```python
print(sorted(data_train['body'].unique()))
```

```python
#В столюце body есть дубликаты вызванные регистром исправим их
data_train['body'] = data_train['body'].str.lower()
#Сделаем это и для data_test
data_test['body'] = data_test['body'].str.lower()
print(sorted(data_train['body'].unique()))
```

```python
print(sorted(data_train['transmission'].unique()))
```

```python
print(sorted(data_train['state'].unique()))
```

```python
print(sorted(data_train['color'].unique()))
```

```python
#для тренровочных данных удалим данные в которых есть прочерк
data_train = data_train.loc[(data_train['color'] != '—') & (data_train['interior'] != '—')]
```

```python
# Избавимся от ненужных в исследовании столбцов
data_train = data_train.drop('vin', axis=1)
vin_test = data_test['vin']
data_test = data_test.drop('vin', axis=1)
```

```python
# объединим марку производителя и модель в один столбец
def union_name(data):
    cols = ['make', 'model', 'trim']
    data['full_auto_name'] = data[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
    data = data.drop(cols, axis=1)
    return data

data_test = union_name(data_test)
data_train = union_name(data_train)
```

```python
display(data_test.head(10))
data_test.info()
display(data_train.head(10))
data_train.info()
```

```python
# Разделим данные на выборки
features_train, target_train = data_train.drop(columns = 'sellingprice', axis=1), data_train['sellingprice']
```

```python
#Закодируем данные Original Encoding
columns_need = ["body","transmission", "state", "color", "interior", "seller", "full_auto_name"]
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
encoder.fit(features_train[columns_need])
features_train[columns_need] = encoder.transform(features_train[columns_need])
display(features_train)
data_test[columns_need] = encoder.transform(data_test[columns_need])
display(data_test)
```

```python
numeric = ['condition', 'odometer', 'car_age']
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
display(features_train)
data_test[numeric] = scaler.transform(data_test[numeric])
display(data_test)
```

```python
features_train['sellingprice'] = target_train
```

```python
#Поссмотрим корреляцию Пирсона и двустороннее значение p в обучающих данных
print(features_train.corr())  

plt.figure(figsize=(15, 10))

# plotting the correlation heatmap  
data_train_plot = sns.heatmap(features_train.corr(), cmap="YlGnBu", annot=True, linewidths = 5)  
    
# displaying the heatmap  
plt.show()
```

```python
#Поссмотрим корреляцию Пирсона и двустороннее значение p в тестовых данных
print(data_test.corr())  

plt.figure(figsize=(15, 10))

# plotting the correlation heatmap  
data_train_plot = sns.heatmap(data_test.corr(), cmap="YlGnBu", annot=True, linewidths = 5)  
    
# displaying the heatmap  
plt.show()
```

```python
features_train = features_train.drop('sellingprice', axis=1)
```

### Обучение моделей

```python
catboost_model = CatBoostRegressor(loss_function = 'MAPE', n_estimators = 100, verbose = 25)
catboost_model.fit(features_train, target_train, plot = True)
```

```python
score = cross_val_score(catboost_model, features_train, target_train, scoring='neg_mean_absolute_percentage_error', cv=5)
print(np.mean(score))
```

```python
params = [{
    'n_estimators': [100, 250, 500],
    'depth': [4, 10,15]
}]

model = GridSearchCV(estimator = CatBoostRegressor(loss_function = 'MAPE',  verbose = 250), param_grid = params,)
model.fit(features_train, target_train)
```

```python
model.best_params_
```

```python
catboost_model_best = CatBoostRegressor(
            loss_function='MAPE',
            n_estimators=1000,
            depth=15,
            verbose=250
)
catboost_model_best.fit(features_train, target_train)
```

```python
#Случайный лес
rf_model = RandomForestRegressor(
    random_state=12345,
    n_estimators=50,
    max_depth=15,
    verbose=False
    )
```

```python
rf_model.fit(features_train, target_train)
```

```python
cv_score_rf = cross_val_score(rf_model, 
                features_train, 
                target_train,
                scoring='neg_mean_absolute_percentage_error',                           
                cv=5)
```

```python
print(cv_score_rf)
```

```python
best_model_predict = catboost_model_best.predict(data_test)
```
